```python
import os
import fitz  # PyMuPDF
import pandas as pd
import re

# Define your list of keywords
keywords = ["farmer", "rancher", "orchards"]

# Function to extract sentences containing keywords
def extract_sentences_with_keywords(pdf_path, keywords):
    doc = fitz.open(pdf_path)
    results = []
    
    for page in doc:
        text = page.get_text("text")
        sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|!)\s', text)  # Splitting into sentences
        
        for sentence in sentences:
            for keyword in keywords:
                if re.search(rf'\b{keyword}\b', sentence, re.IGNORECASE):
                    results.append((os.path.basename(pdf_path), keyword, sentence.strip()))
    
    return results

# Folder containing PDF files
pdf_folder = r"C:\Users\shaun.rolph\USDA\AMS Grants Division (GD) - ODA\Flattened Narratives"  # Folde  # Change this to your folder path
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith(".pdf")]

data = []
for pdf in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf)
    data.extend(extract_sentences_with_keywords(pdf_path, keywords))

# Convert to DataFrame
df = pd.DataFrame(data, columns=["PDF Document Name", "Keyword Found", "Sentence Containing Keyword"])

# Output file location
output_csv_path = r"C:\Users\shaun.rolph\USDA\AMS Grants Division (GD) - ODA\Flattened Narratives\keywords_DBIG.csv"  # Change this to your desired output location
df.to_csv(output_csv_path, index=False)

# Display the table
df
```
